

*   Importe de la liberia para que se puedan mostrar los nombres de los  meses en lugar de los numeros con lo que se encuentran representados en la base de datos



```python
import calendar

# Agrupar los datos por célula terrorista y mes, y contar los ataques
grouped_data = filtered_data.groupby(['gname', 'imonth']).size().reset_index(name='attack_count')

# Obtener las células terroristas más frecuentes (puedes ajustar este número según tus necesidades)
top_cells = grouped_data['gname'].value_counts().nlargest(7).index

# Iterar sobre cada célula terrorista y generar un gráfico individual
for cell in top_cells:
    # Filtrar los datos para la célula terrorista actual
    cell_data = grouped_data[grouped_data['gname'] == cell]

    # Crear una nueva figura y ejes para el gráfico actual
    fig, ax = plt.subplots(figsize=(8, 6))

    # Obtener los nombres abreviados de los meses
    month_labels = [calendar.month_abbr[m] for m in cell_data['imonth']]

    # Trazar la serie de tiempo de ataques para la célula terrorista actual
    ax.plot(month_labels, cell_data['attack_count'])

    # Configurar etiquetas y título del gráfico actual
    ax.set_xlabel('Mes')
    ax.set_ylabel('Cantidad de ataques')
    ax.set_title(cell)

    # Mostrar el gráfico actual
    plt.show()
```

```python
# Agrupar los datos por célula terrorista y año, y contar los ataques
grouped_data = filtered_data2.groupby(['gname', 'iyear']).size().reset_index(name='attack_count')

# Obtener las células terroristas más frecuentes (puedes ajustar este número según tus necesidades)
top_cells = grouped_data['gname'].value_counts().nlargest(10).index

# Iterar sobre cada célula terrorista y generar un gráfico individual
for cell in top_cells:
    # Filtrar los datos para la célula terrorista actual
    cell_data = grouped_data[grouped_data['gname'] == cell]

    # Crear una nueva figura y ejes para el gráfico actual
    fig, ax = plt.subplots(figsize=(8, 6))

    # Trazar la serie de tiempo de ataques para la célula terrorista actual
    ax.plot(cell_data['iyear'], cell_data['attack_count'])

    # Configurar etiquetas y título del gráfico actual
    ax.set_xlabel('Año')
    ax.set_ylabel('Cantidad de ataques')
    ax.set_title(cell)  # Utilizar el nombre de la célula terrorista como título

    # Mostrar el gráfico actual
    plt.show()
```
